```python
import json

def extract_names(keywords):
    try:
        # Ensure the input is a string
        keywords = str(keywords)
        
        # Check if the string looks like a JSON array
        if keywords.strip().startswith('[') and keywords.strip().endswith(']'):
            keywords_list = json.loads(keywords)
            # Extract 'name' from each dictionary if it exists
            return [item['name'] for item in keywords_list if 'name' in item]
        else:
            # If it's not a valid JSON array, return an empty list
            return []
    except (json.JSONDecodeError, TypeError) as e:
        print(f"Error decoding JSON: {e}")  # Debugging line to print errors
        return []

# Apply the function to the 'keywords' column
movies['keywords'] = movies['keywords'].apply(extract_names)

# Check the result
print(movies['keywords'].head())
```

```python
import json
import pandas as pd

def extract_top_3_cast(cast_str):
    try:
        # Print cast_str to debug its structure
        print(cast_str)  # Remove this in production after debugging
        
        if not cast_str or cast_str == '[]' or pd.isna(cast_str):
            return []
        
        # Check if it's a string that needs conversion to JSON
        if isinstance(cast_str, str):
            cast_list = json.loads(cast_str)  # Convert string to list of dicts
        else:
            return []

        # Get the top 3 names
        names = [member['name'] for member in cast_list[:3]]
        return names
    except Exception as e:
        print(f"Error: {e}")  # Print error message for debugging
        return []

# Apply function to the 'cast' column
movies['cast'] = movies['cast'].apply(extract_top_3_cast)

# Show the result
print(movies[['title_x', 'cast']].head())
```

```python
import pandas as pd
import json

# Assuming df is your DataFrame and it contains a column 'crew' with a JSON string or list
# If the 'crew' column is in JSON string format, we need to load it first

# Check if 'crew' column contains strings (if not, it's already a list of dictionaries)
if isinstance(movies['crew'][0], str):  # If it's a JSON string
    movies['crew'] = movies['crew'].apply(json.loads)  # Convert to list of dicts

# Now we extract the directors
def get_directors(crew):
    # Filter the list of crew members for 'Director' job
    directors = [member['name'] for member in crew if member['job'] == 'Director']
    return directors

# Apply the function to the 'crew' column and store the result in a new column
movies['directors'] = movies['crew'].apply(get_directors)

# Show the result
print(movies[['directors']])
```

```python
# Assuming 'movies' is your DataFrame

# Function to clean up each column
def clean_column(column):
    # Join the list elements into a single string, remove spaces, and convert to lowercase
    return column.apply(lambda x: ''.join([str(i).lower() for i in x]))

# Apply to the relevant columns in the 'movies' DataFrame


movies['cast'] = clean_column(movies['cast'])
movies['directors'] = clean_column(movies['directors'])



# Display the cleaned DataFrame
print(movies[[ 'cast', 'directors']])
```

```python
# Convert the genres column to lowercase and then join the list into a string
movies['genres'] = movies['genres'].apply(lambda x: ' '.join([genre.lower() for genre in x]) if isinstance(x, list) else '')
movies['keywords'] = movies['keywords'].apply(lambda x: ' '.join([genre.lower() for genre in x]) if isinstance(x, list) else '')
movies['overview'] = movies['overview'].apply(lambda x: ' '.join([genre.lower() for genre in x]) if isinstance(x, list) else '')
movies['tagline'] = movies['tagline'].apply(lambda x: ' '.join([genre.lower() for genre in x]) if isinstance(x, list) else '')
# Show the result
print(movies[['title_x', 'genres']].head())
```

```python
import pandas as pd
import re
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS  # To use built-in stopwords

# Initialize Lemmatizer
lemmatizer = WordNetLemmatizer()

# Step 1: Clean, Split, Lemmatize, and Remove Stopwords
def clean_split_lemmatize(text):
    if isinstance(text, str):
        text = text.replace(',', ' ')  # Replace commas with space
        words = re.findall(r'[a-zA-Z]+', text.lower())  # Extract words
        lemmatized_words = [lemmatizer.lemmatize(word) for word in words if word not in ENGLISH_STOP_WORDS]  # Lemmatize and remove stopwords
        return ' '.join(lemmatized_words)  # Return as clean string
    else:
        return ''

# Step 2: Apply to your column
movies['tagline'] = movies['tagline'].apply(clean_split_lemmatize)

# Step 3: Create CountVectorizer with stopwords
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(movies['tagline']).toarray()

# Step 4: Check output
print(vectors.shape)
print(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=movies[movies['title_x'] == movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:6]

    for i in movie_list:
        print(movies.iloc[i[0]].title_x)
        


    
```

```python
recommend('Spider-Man 3')
```
